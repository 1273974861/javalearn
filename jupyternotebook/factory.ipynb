```python
import numpy as np
import pandas as pd

csv1 = pd.read_csv('E:\\yuchulidata\\production_missing_num.csv')
print(csv1)
# 数据概况的观察
print(csv1.describe())
print("\n\n\n\n")
df = csv1
```

```python
# 属性字段数值化
df = df.replace(['A', 'B', 'C', 'D', 'E'], [1, 2, 3, 4, 5])
df = df.replace("FALSE", 0)
df = df.replace("TRUE", 1)
column2 = df

print(column2)
print("已知第24行为缺失数值的行")
print(column2[24:25])
# 缺失值的填充或删除
# 把缺失值中的None转换成np.NAN格式,str类型转为float 方便处理
column2 = column2.replace("None", np.NAN)
column2 = column2.astype(float)
print("处理结果")
column2
```

```python
# 使用常数填充
# 将所有的None换成0
column3 = column2.fillna(0)
print("使用常数法填充后的24行")
print(column3[24:25])
# 统计缺省值NaN的个数
column2.isnull().sum()
```

```python
# 求这一行的聚合值
column2_mean = column2.mean()
print(column2_mean)
# 使用平均数填充
column3 = column2.fillna(column2_mean)
print("使用平均数填充后的24行")
print(column3[24:25])
```

```python
column2_mid = column2.median()
print(column2_mid)
# 使用中位数填充
column3 = column2.fillna(column2_mid)
print(column3[24:25])
```

```python
zscore = (column3 - column3.mean()) / column3.std()
zscore_r = zscore[abs(zscore) <= 3]
print(zscore_r)

normalized = zscore_r / (zscore_r.max() - zscore_r.min())
print("____________________")
normalized["fault_flg"] = column2["fault_flg"]
normalized
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)

normalized = normalized.dropna()
normalized.isna().sum()
pca.fit_transform(normalized)
```
