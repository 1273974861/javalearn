```python
#数值化 对production
production = production.replace(['A', 'B', 'C', 'D', 'E'], [1, 2, 3, 4, 5])
a = production['fault_flg']
a = pd.DataFrame(a)
for i in a.columns:
    if a[i].dtype==bool:
        a[i]=a[i].astype('int')
production['fault_flg'] = a
production
```

```python
#数值化 对production_missing_category
production_missing_category = production_missing_category.replace(['A', 'B', 'C', 'D', 'E'], [1, 2, 3, 4, 5])
a = production_missing_category['fault_flg']
a = pd.DataFrame(a)
for i in a.columns:
    if a[i].dtype==bool:
        a[i]=a[i].astype('int')
production_missing_category['fault_flg'] = a
production_missing_category
```

```python
#数值化 对production_missing_category
production_missing_num = production_missing_num.replace(['A', 'B', 'C', 'D', 'E'], [1, 2, 3, 4, 5])
a = production_missing_num['fault_flg']
a = pd.DataFrame(a)
for i in a.columns:
    if a[i].dtype==bool:
        a[i]=a[i].astype('int')
production_missing_num['fault_flg'] = a
production_missing_num
```

```python
#数值化 production_missing_num_4_redshift
production_missing_num_4_redshift = production_missing_num_4_redshift.replace(['A', 'B', 'C', 'D', 'E'], [1, 2, 3, 4, 5])
a = production_missing_num_4_redshift['fault_flg']
a = pd.DataFrame(a)
for i in a.columns:
    if a[i].dtype==bool:
        a[i]=a[i].astype('int')
production_missing_num_4_redshift['fault_flg'] = a
production_missing_num_4_redshift
```

```python
##归一化
#定义01规一的方法
def guiyi(x):
    MAX = np.max(x)
    MIN = np.min(x)
    x1 = (x-MIN)/(MAX-MIN)
    return x1
#production归一化
a = production['type']
b = production['length']
c = production['thickness']
d = production['fault_flg']
production['type'] = guiyi(a)
production['length'] = guiyi(b)
production['thickness'] = guiyi(c)
production['fault_flg'] = guiyi(d)
production
```

```python
#production_missing_category归一化
a = production_missing_category['type']
b = production_missing_category['length']
c = production_missing_category['thickness']
d = production_missing_category['fault_flg']
production_missing_category['type'] = guiyi(a)
production_missing_category['length'] = guiyi(b)
production_missing_category['thickness'] = guiyi(c)
production_missing_category['fault_flg'] = guiyi(d)
production_missing_category
```

```python
#production_missing_num_4_redshift归一化
a = production_missing_num_4_redshift['type']
b = production_missing_num_4_redshift['length']
c = production_missing_num_4_redshift['thickness']
d = production_missing_num_4_redshift['fault_flg']
production_missing_num_4_redshift['type'] = guiyi(a)
production_missing_num_4_redshift['length'] = guiyi(b)
production_missing_num_4_redshift['thickness'] = guiyi(c)
production_missing_num_4_redshift['fault_flg'] = guiyi(d)
production_missing_num_4_redshift
```

```python
#对hotel进行降维处理，4变2维
from sklearn.decomposition import PCA
pca = PCA(n_components=2)

production = production.dropna()
a = pca.fit_transform(production)
a
```

```python
#散点图绘制
import matplotlib.pyplot as plt
plt.scatter(a[:,0],a[:,1])
plt.xlabel('xzhou')
plt.ylabel('yzhou')
plt.title('production')
plt.show()
```
