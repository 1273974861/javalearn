```python
#读取三个txt文件
import csv
nahan = []
with open('E://yuchulidata//txt//nahan.txt', 'rt') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        nahan.append(row)
nahan2 = [str(i) for i in nahan]
nahan3 = ''.join(nahan2)
nahan3
```

```python
shadow = []
with open('E://yuchulidata//txt//shadow.txt', 'rt') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        shadow.append(row)
print(shadow)
shadow = [str(i) for i in shadow]
shadow = ''.join(shadow)
print(shadow)
```

```python
yukikuni = []
with open('E://yuchulidata//txt//yukikuni.txt', 'rt') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        yukikuni.append(row)
print(yukikuni)
yukikuni = [str(i) for i in yukikuni]
yukikuni = ''.join(yukikuni)
print(yukikuni)
```
