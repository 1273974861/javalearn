```python
#数值化处理
#根据两组数据都已经是数值无须此步骤
#进行归一化
#查看两数据后对第二组数据进行归一化
#先删去其year_month列
monthly = monthly.drop(['year_month'],axis=1)
def guiyi(x):
    MAX = np.max(x)
    MIN = np.min(x)
    x1 = (x-MIN)/(MAX-MIN)
    return x1
a = monthly['sales_amount']
b = monthly['customer_number']
monthly['sales_amount'] = guiyi(a)
monthly['customer_number'] = guiyi(b)
monthly
```

```python
#可视化
import matplotlib.pyplot as plt
plt.scatter(monthly['sales_amount'],monthly['customer_number'])
plt.xlabel('xzhou')
plt.ylabel('yzhou')
plt.title('sandiantu')
plt.show()
```
