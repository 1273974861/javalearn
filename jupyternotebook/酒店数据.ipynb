```python
##归一化
#定义01规一的方法
def guiyi(x):
    MAX = np.max(x)
    MIN = np.min(x)
    x1 = (x-MIN)/(MAX-MIN)
    return x1

#对customer中的三列进行归一
a = customer['age']
b = customer['home_latitude']
c = customer['home_longitude']
customer['age'] = guiyi(a)
customer['home_latitude'] = guiyi(b)
customer['home_longitude'] = guiyi(c)
customer
```

```python
#对酒店中的数据进行归一
a = hotel['base_price']
b = hotel['hotel_latitude']
c = hotel['hotel_longitude']
d = hotel['is_business']
hotel['base_price'] = guiyi(a)
hotel['hotel_latitude'] = guiyi(b)
hotel['hotel_longitude'] = guiyi(c)
hotel['is_business'] = guiyi(d)
hotel
```

```python
#对holiday进行归一
a = holiday['holidayday_flg']
b = holiday['nextday_is_holiday_flg']
holiday['holidayday_flg'] = guiyi(a)
holiday['nextday_is_holiday_flg'] = guiyi(b)
holiday
```

```python
#对reserve中的price归一化
reserve
a = reserve['total_price']
reserve['total_price'] = guiyi(a)
reserve
```
